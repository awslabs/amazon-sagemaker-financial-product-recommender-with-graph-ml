# Download data from Harvard and SEC websites

```python
!mkdir -p source_data
for year in range(2004, 2017+1):
    for q in range(1, 4+1):
        fn = '13f-list.pdf' if (year, q) == (2004, 1) else f'13f/13flist{year}q{q}.pdf'
        url = f'https://www.sec.gov/divisions/investment/{fn}'   
        !wget {url} -O source_data/{year}q{q}.pdf
```
